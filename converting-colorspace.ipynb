# Converting Color Space

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def img_show(image):
    cvt_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(cvt_img)
```

```python
file = "test.jpeg"
```

## RGB to Grey

### Read file as greyscale

```python
img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
```

```python
img_show(img)
```

### Convert color space

```python
img2 = cv2.imread(file, cv2.IMREAD_COLOR)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
img_show(img2)
```

## Brightness 

dst(x,y) = src(x,y) + n

```python
img = cv2.imread(file)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = img + 100
```

```python
plt.imshow(img)
```

```python
img = cv2.imread(file, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j] = img[i,j] + 100

plt.imshow(img)
```

```python
img = cv2.imread(file, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j] = img[i,j] + 100
        if img[i,j].any() > 255 : img[i,j] = 255
        elif img[i,j].any() < 0 : img[i,j] = 0

plt.imshow(img)
```

## RGB space to HSV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "../dataset/sesumi.jpeg"

img = cv2.imread(file)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_red = np.array([50,50,50])
upper_red = np.array([130,255,255])

mask = cv2.inRange(hsv_img, lower_red, upper_red)

res = cv2.bitwise_and(img, img, mask=mask)


plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("image")
plt.axis('off')
plt.show()
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.title("mask")
plt.axis('off')
plt.show()
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
plt.title("result")
plt.axis('off')
plt.show()
```
