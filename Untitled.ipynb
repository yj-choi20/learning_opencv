## Image Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../dataset/gradient_img.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
retVal, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
retVal, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
retVal, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
retVal, thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
retVal, thresh5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
```

```python
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
